```python
# =========================
# Rule Base
# =========================

rules = [
    ({"fever", "cough"}, "Flu"),
    ({"fever", "rash"}, "Dengue"),
    ({"headache", "nausea"}, "Migraine"),
    ({"chest_pain", "breathlessness"}, "Heart Issue"),
    ({"fatigue", "weight_loss"}, "Diabetes")
]
```

```python
# =========================
# Forward Chaining Engine
# =========================

facts = set()
logs = []

def add_fact(symptom):
    facts.add(symptom)
    logs.append(f"Fact added: {symptom}")

def diagnose():
    diagnosis = None

    for conditions, result in rules:
        logs.append(f"Checking rule {conditions} -> {result}")

        if conditions.issubset(facts):
            logs.append("Rule matched ✅")
            diagnosis = result
        else:
            logs.append("Rule not matched ❌")

    return diagnosis
```

```python
# =========================
# User Input Section
# =========================

print("Enter symptoms (type done to stop):")

while True:
    s = input("Symptom: ").lower()

    if s == "done":
        break

    add_fact(s)
```

```python
result = diagnose()

print("\n===== Diagnosis =====")

if result:
    print("Disease:", result)
else:
    print("No disease matched")
```

```python
print("\n===== Reasoning Steps =====")

for step in logs:
    print(step)
```
